## Now we improve the accuracy

```python
path = dt_model1.cost_complexity_pruning_path(x1 , y1)  ## Post Pruning(remove the leaf node(branch) to avoid overfitting)
path
```

```python
ccp_alphas = path["ccp_alphas"]
ccp_alphas  ## initial ccp_alpha starts with 0.0
```

```python
dt_model2 = []
score = []
for ccp in ccp_alphas:
    count = 1
    dt_m = DecisionTreeClassifier(ccp_alpha= ccp)
    dt_m.fit(x1 , y1)
    dt_model2.append(dt_m)
    score.append(dt_m.score(X_test , y_test))
    
```

```python
dt_model2 ## we build multiple model with different ccp_alphas
```

```python
score
```

for ex In 2nd ccp value 0.06666666666666667 ; so its like threshould anything greater than this value it leave the branch 

## now we visualize

```python
plt.plot(score , "r-+")
plt.xlabel("Iteration")
plt.ylabel("Scores")
```

```python
train_scores = [i.score(x1,y1) for i in dt_model2]
train_scores
```

```python
test_scores = [i.score(X_test , y_test) for i in dt_model2]
test_scores
```

```python
fig , ax =plt.subplots()
ax.set_xlabel("Alpha")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy vs Alpha for training  and Testing sets")
ax.plot(ccp_alphas , train_scores , marker = "o" , label ="train" , drawstyle = "steps-post")
ax.plot(ccp_alphas ,test_scores , marker = "o" , label ="test" , drawstyle = "steps-post")
ax.legend()
plt.show()
```

**In 0.30 Alpha the test and train accuracy distance is low compare to others train acc-0.7 test acc-45 becozz insufficient data**

## hyperparameter tunning
- to build the all the paramter optimized we use hyperparameter tunning
- we use **Randomsearch or Gridsearch CV**

Grid and Random both are same the only difference is grid more effective and computation and random is not that much

```python
grid_parm = {
             "criterion":['gini','entropy'],
             "splitter":['best','random'],
             "max_depth" : range(2,40,1),
             "min_samples_split":range(2,10 ,1),
             "min_samples_leaf":range(1,10,1),
             'ccp_alpha':np.random.rand(20)
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=dt_model_ccp , param_grid=grid_parm , cv=5 , n_jobs=-1)
```

grid.fit(x1 , y1)  ## it takes more time based on the system

```python
grid.best_params_
```
