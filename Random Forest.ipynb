# Random Forest
Random Forest is a supervised learning .it is used to solve both classification and regression problem

Random forest is the algorithm used for both classification and regression problem. it works with the form of Ensemble Technique....Ensemble Technique helps to combine multiple model and predict the result

## Bootstrap & Aggregation
Random Forest is the form of Bootstrap & Aggregation ... where it create many Decisition tree and predict the result ... The main difference from Decision Tree & Random Forest ..


- Decision Tree has Low Bias & High Variance
- in Random Forest has low Bias and Low Variance becozz it combines all the decision tree in the model so high variance converted into low variance

```python
## Today we solve the inbuild dataset from sklearn
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
# these are the properties of the dataset
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
# we see the first 4 digits from our datset it is 8*8 matrix
```

```python
digits.data
# this is our data column from digits
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
df["target"]=digits.target
```

```python
df.head()
# so here 0 to 63 is our independent (data)
# target is our dependet data
```

```python
df.shape
```

```python
# seperate our dependent and independent data as x and y
x=df.drop(["target"],axis=1)
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
predict=model.predict(X_test)
```

```python
## Now we see the confussion matrix
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,predict)
cm
```

```python
## Now we see the seaborn 
import seaborn as sns 
sns.heatmap(cm,annot=True)
plt.xlabel=("predicted")
plt.ylabel=("truth")
```
